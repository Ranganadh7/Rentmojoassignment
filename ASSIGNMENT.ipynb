```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_excel('C:/Users/hp/Desktop/RENTMOJO/data.xlsx')
df.head(10)
```

```python
# 1ST QUERY
df['runtime'] = df['runtime'].replace(0,np.NAN)
df['runtime'].min()
```

```python
def find_minmax(x):
    #use the function 'idmin' to find the index of lowest profit movie.
    min_index = df[x].idxmin()
    #use the function 'idmax' to find the index of Highest profit movie.
    high_index = df[x].idxmax()
    high = pd.DataFrame(df.loc[high_index,:])
    low = pd.DataFrame(df.loc[min_index,:])
    
    #print the movie with high and low profit
    print("Movie Which Has Highest "+ x + " : ",df['title'][high_index])
    print("Movie Which Has Lowest "+ x + "  : ",df['title'][min_index])
    return pd.concat([high,low],axis = 1)
find_minmax('runtime')
```

```python
# 2nd QUERY
df.sort_values(by=('budget'),inplace=True,ascending=False)
df[['title','vote_average','revenue']].head(5)
```

```python
# 2nd QUERY:-

#Take the top 5 production houses (by budget) and list their top 5 most popular movies, their revenue and vote_average
def count_genre(x):
    #concatenate all the rows of the genrs.
    data_plot = df[x].str.cat(sep = '|')
    data = pd.Series(data_plot.split('|'))
    #conts each of the genre and return.
    info = data.value_counts(ascending=False)
    return info
pc= count_genre('production_companies')
print(pc)
print(pc.iloc[:5])
```

```python
# 3rd QUERY:-
 #List the production house for every year from 2000-2016 which has released the most number of movies in that year 
production_companies= count_genre('production_companies')

#plot he barh plot.
production_companies.iloc[:20].plot(kind='barh',figsize=(16,8),fontsize=13)
plt.title("Production Companies Vs Number Of Movies",fontsize=15)
plt.xlabel('Number Of Movies',fontsize=13)
plt.ylabel("Production House",fontsize=13)
plt.yticks=[]
sns.set_style("whitegrid")
```
